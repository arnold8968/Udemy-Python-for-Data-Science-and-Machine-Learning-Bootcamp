```python
import numpy as np

x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x, z)
ax1.set_xlabel('x')


ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x, z)
ax2.set_xlim([20, 22])
ax2.set_ylim([300, 500])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```
