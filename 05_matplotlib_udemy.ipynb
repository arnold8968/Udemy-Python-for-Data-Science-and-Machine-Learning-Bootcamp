```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Functional

plt.plot(x, y)
```

```python
plt.plot(x, y, 'r-')
```

```python
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.3,0.4,0.8])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
'''
Figure Size and DPI
'''
```

```python
fig = plt.figure(figsize=(3,2), dpi=100) # dpi means the pixel the plot. Ususlly use the default value

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8,2)) # dpi means the pixel the plot. Ususlly use the default value

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =1, figsize=(8,2)) # dpi means the pixel the plot. Ususlly use the default value

axes[0].plot(x, y)
axes[1].plot(y, x)

fig.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi = 200)
```

```python
fig = plt.figure() # dpi means the pixel the plot. Ususlly use the default value

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x ** 2, label='X Squared')
ax.plot(x, x ** 3, label='X Cubed')

ax.legend() # ax.plot  label
```

```python
'''
Part 3
Plot Appearance
'''
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='green') # 'FF8C00' RGB Hex Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', linewidth = 20)  # linewidth default is 1
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', linewidth = 20, alpha = 0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', linewidth = 20, alpha = 0.5, linestyle=':') # linestyle '--'
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', lw = 20, alpha = 0.5, ls=':') # linestyle = ls; linewidth = lw
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', lw = 1, alpha = 0.5, ls='-', marker='o', markersize = 2)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', lw = 1, alpha = 0.5, ls='-', marker='o', markersize = 10,
       markerfacecolor = 'yellow')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', lw = 1, ls='-')

ax.set_xlim([0, 1])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='purple', lw = 1, ls='-')

ax.set_xlim([0, 1])
ax.set_ylim([0, 5])
```
