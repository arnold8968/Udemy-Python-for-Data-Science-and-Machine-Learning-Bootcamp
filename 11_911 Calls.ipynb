```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./dataset/911.csv')
df.head()
```

```python
df.info()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
len(df['title'].unique())
```

```python
df['title'][0].split(':')[0]
```

```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

```python
df.head()
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x = 'Reason', data=df)
```

```python
type(df['timeStamp'][0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df.head()
```

```python
df['timeStamp'][0].month
```

```python
dmap = {0: 'Mon', 1: 'Tue',2:'Web',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['timeStamp'][0].weekday()
```

```python
dmap[df['timeStamp'][0].weekday()]
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Day of Week'] = df['timeStamp'].apply(lambda x: dmap[x.weekday()])
```

```python
df.head()
```

```python
sns.countplot(x='Day of Week', hue='Reason', data=df)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1)
```
